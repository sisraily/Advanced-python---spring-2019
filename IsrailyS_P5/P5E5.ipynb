Project 5-5: Finding words containing ss

In this exercise, we'll search a file for words containing "ss" and printing every one of them. To start, we import the regular expression library and open the file.

```python
import re

wordlist = open('wordlist.txt', 'r')
line = wordlist.readline
```

After stripping the whitespace characters from the file using the rstrip method, we compare each word to a regular expression. The regular expression we're using contains two lowercase or capital Ss between 0 or more non-whitespace characters. Every word matching the regular expression is printed if it is more than 0 characters long.

```python
for line in wordlist:
    words = line.rstrip()
    SSword = re.findall('\S*[sS][sS]\S*', words)
    if len(SSword) > 0:
        print(SSword)
```
