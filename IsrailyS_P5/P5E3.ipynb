Project 5-3: Finding words ending with s

In this exercise, we'll search a file for words ending with s. To start, we import the regular expression library, open the file, and prepare it for reading.

```python
import re

wordlist = open('wordlist.txt', 'r')
line = wordlist.readline
```

First, we run rstrip on the words to remove the whitespace characters. Now that the words are separated, we use a regular expression which contains any number of nonwhitespace characters followed by a capital or lowercase S. We use non-whitespace characters instead of letters in special cases like "x-axis" which use characters other than letters. The dollar sign at the end means "s" is at the end of the word. We test if each word is longer than 0 characters and print if true.

```python
for line in wordlist:
    words = line.rstrip()
    wordING = re.findall('\S*[sS]$', words)
    if len(wordING) > 0:
        print(wordING)
```
