Project 5-4: Finding words ending with ing

In this exercise, we'll search a file for words ending with ing. To start, we import the regular expression library, open the file, and prepare it for reading. We also create an empty list 

```python
import re

wordlist = open('wordlist.txt', 'r')
line = wordlist.readline
```

First, we run rstrip on the words to remove the whitespace characters. Now that the words are separated, we use a regular expression which contains any number of non-whitespace characters followed by "ing". We use letters other than characters since some words contain special characters like dashes. The dollar sign at the end means "ing" is at the end of the word. We test if each word is longer than 0 characters and print if true.

```python
for line in wordlist:
    words = line.rstrip()
    wordING = re.findall('\S+ing$', words)
    if len(wordING) > 0:
        print(wordING)
```
