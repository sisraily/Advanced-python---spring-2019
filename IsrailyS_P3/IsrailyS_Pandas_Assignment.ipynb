### Histogram Chart

I created a new dataframe to better show an example of histogram graph. 
The dataframe, called 'af, contains a dictionary with two arrays. Using numpy's random.randn function, 
each array is programed to holds 500 random standard distribution numbers.

```python
af = pd.DataFrame({"Standard number distribution 'A'": np.random.randn(500), "Standard number distribution 'B'":np.random.randn(500)})

```

##### Pandas plot function explained:

The plot function within pandas is used to draw the histogram. 
The first parameter, 'kind', takes in the type of graph ('hist' coresponds to histogram).
The 'figsize' parameter takes in a tuple of the width and hight of the graph in inches.
The 'edgecolor' parameter corresponds to the edge color of the bars. This is set to 'w', for white.
The 'subplots' parameter takes in a bool value; when set to True the two arrays are graphed separately. 
Lastly, the 'title' parameter of the graphs are also set and that takes in a string.


```python
af.plot(kind='hist',figsize=(10, 10), edgecolor='w', subplots=True, title='Histogram chart of 500 numbers in standard distribution')
```
